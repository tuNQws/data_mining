```python
def caught_speeding(speed, is_birthday):
    if is_birthday:
        speed -= 5
    if speed <= 60:
        return "No Ticket"
    elif speed >= 61 and speed <= 80:
        return "Small Ticket"
    else:
        return "Big Ticket"
```

```python
print(caught_speeding(81, True))  # "Small Ticket"
print(caught_speeding(81, False))  # "Big Ticket"
```

```python
print(caught_speeding(81, True))
```
