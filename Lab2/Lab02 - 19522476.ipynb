```python
#2 Job Market
#Load the data using Pandas
import pandas as pd
data = pd.read_csv('job-market (1).csv')
#Visualize top 10 first rows
data.head(10)
```

```python
data.info()
data.dropna(inplace=True)
data.drop_duplicates(inplace=True)
```

```python
#Fix column datatypes
import re

def clean_string(text):
    if text:
        # Remove HTML tags
        clean = re.compile('<.*?>')
        text = re.sub(clean, '', text)       
        # Replace ' ' and '\n * ' with a space
        text = re.sub(r'( |\n * )', ' ', text)        
        # Remove unwanted characters
        text = re.sub(r'\n *|\*', '', text)
        
        return text if text else None
```

```python
data['Id'] = data['Id'].astype(int)
data['Title'] = data['Title'].astype(str)
data['Company'] = data['Company'].astype(str)
data['Location'] = data['Location'].astype(str)
data['Area'] = data['Area'].astype(str)
data['Classification'] = data['Classification'].astype(str)
data['SubClassification'] = data['SubClassification'].astype(str)
data['Requirement'] = data['Requirement'].astype(str)
data['LowestSalary'] = data['LowestSalary'].astype(int)
data['HighestSalary'] = data['HighestSalary'].astype(int)
data['Date'] = pd.to_datetime(data['Date'], format='%Y-%m-%dT%H:%M:%S.%fZ')
data['LowestSalary'] = pd.to_numeric(data['LowestSalary'], errors='coerce')
data['HighestSalary'] = pd.to_numeric(data['HighestSalary'], errors='coerce')
data['FullDescription'] = data['FullDescription'].apply(clean_string)
data['JobType'] = data['JobType'].astype(str)
```

```python
import re

# Load the sales.csv file into a DataFrame
df = pd.read_csv('sales.csv')

# Define a regular expression pattern to match the month and year portion of the ordered_at column
pattern = re.compile(r'(\d{4})-(\d{2})-\d{2}')

# Use a lambda function with the regular expression pattern to extract the month and year
df['month_year'] = df['ordered_at'].apply(lambda x: '-'.join(re.match(pattern, x).group(2, 1)))

# Print the DataFrame with the new month_year column
print(df.head())
```
